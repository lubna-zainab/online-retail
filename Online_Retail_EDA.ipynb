```python
#Getting basic info of our data: columns we have, the count of non null entries and the data types.
df.info()
```

```python
# we can see that all columns have appropriate data types except for InvoiceDate which should be a datetime datatype but is an object instead.
# Although CustomerID column having float values is okay, it's better to convert it to int as 
# If we observe carefully, we can see not all columns have non null values. We can see this in detail further.
```

```python
#Let us first change the data type of InvoicecDate column

pd.to_datetime(df['InvoiceDate'])
```

```python
# We can see that the dtype has been converted to datetime
```

```python
# checking count of null/missing value columns that we have.
df.isnull().sum()

# Since CustomerID column has null values, we will change the datatype later 
```

```python
# We will further see the columns that are null in detail
df[df["CustomerID"].isnull()]
```

```python
# Customer ID is a very important feature and we cannot impute it with mean, median, mode or other imputing methods in this case, Hence, it is better to drop the columns.
```

```python
df[df["Description"].isnull()]
```

```python
# We can see that all null values in Description have null values in Customer ID as well. Thus, if we drop null values in Customer ID column, the null values in Description will also drop.
```

```python
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df["CustomerID"]=df["CustomerID"].astype(int)
```
